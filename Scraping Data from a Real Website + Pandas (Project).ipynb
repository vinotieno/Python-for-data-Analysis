```python
## We gonna start by Importing pandas, and beautifulsoup 

## We need the url of the website

## using the requests.get() to get the url 

## soup to pull out the information
```

```python
import requests

from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/Counties_of_Kenya'



page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

print(soup)
```

```python
table =  soup.find_all('table')[2]

print(table)


```

```python
# Lets pull out the 'th' tags in our table 
```

```python
table.find_all('th')
```

```python
Counties = table.find_all('th')
```

```python
print(Counties)
```

```python
Counties = table.find_all('th')[0:8]

print(Counties)


```

```python
## Now using List Comprehension---- [i for i in iteration if expression]
```

```python
Counties = [title.text for title in Counties]

print(Counties)
```

```python
#### Using strip() to clean our data
```

```python
Counties = [title.strip() for title in Counties]

print(Counties)
```

```python
####  Now we are gonna Import our pandas now using the package pd to visualize our table in python
```

```python
import pandas as pd


df = pd.DataFrame(columns = Counties)

df



```

```python
# table row and data 
```

```python
column_data = table.find_all('tr')

for row in column_data:
    print(row.find_all('td'))
```

```python
# Using list comprehension to get the row data and entire data
```

```python
column_data = table.find_all('tr')
```

```python


for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)

```

```python
## To remove the first empty list we use index
```

```python
for row in column_data[1:48]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
```

```python
# There our final project is here. 


# We extracted a list of 47 counties in kenya and exported it to csv
```

```python
import pandas as pd


df = pd.DataFrame(columns = Counties)

df

for row in column_data[1:48]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    
    length = len(df)
    df.loc[length] = individual_row_data

df
```

```python
# Export our data 
```
