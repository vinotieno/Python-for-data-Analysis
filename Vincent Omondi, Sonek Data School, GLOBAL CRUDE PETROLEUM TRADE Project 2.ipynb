```python
top_10_countries = df.nlargest(10, 'Trade Value')
print(top_10_countries[['Country', 'Trade Value']])
```

# B.	Make 5 unique visualizations of your own choosing for the different columns of your dataset. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Bar Plot of the Top 10 Countries by Tourism Revenue

```python
# Bar plot of top 10 countries by trade value
plt.figure(figsize=(11, 4))
sns.barplot(x='Trade Value', y='Country', data=top_10_countries)
plt.title('Top 10 Countries by Trade Value')
plt.xlabel('Trade Value (in millions)')
plt.ylabel('Country')
plt.show()

# Explanation: This bar plot shows the top 10 countries based on their trade value.
```

# 2.Pie Chart of the Top 10 Countries by Tourism Revenue

```python
# Pie chart of top 10 countries by Trade Value
plt.figure(figsize=(10, 10))
plt.pie(top_10_countries['Trade Value'], labels=top_10_countries['Country'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', 10))
plt.title('Distribution of Trade Value Among Top 10 Countries')
plt.show()

# Explanation: This pie chart visualizes the distribution of Trade Value among the top 10 countries.
```

# 3. Box Plot of Tourism Revenue

```python
# Box plot of Trade Value
plt.figure(figsize=(8, 6))
sns.boxplot(x='Trade Value', data=df, color='lightblue')
plt.title('Box Plot of Trade Value')
plt.xlabel('Trade Value (in millions)')
plt.show()

# Explanation: This box plot displays the distribution of Trade Value across all countries, highlighting the spread and potential outliers.
```

# 4. Histogram of Trade Value

```python
# Histogram of Trade Value
plt.figure(figsize=(10, 6))
sns.histplot(df['Trade Value'], bins=15, kde=True, color='blue')
plt.title('Histogram of Trade Value')
plt.xlabel('Trade Value (in millions)')
plt.ylabel('Frequency')
plt.show()

# Explanation: This histogram shows the frequency distribution of Trade Value, providing insights into its overall distribution.
```

# 5. Scatter Plot of Trade Value vs. Country

```python
# Scatter plot of Trade Value vs. country
plt.figure(figsize=(15, 6))
sns.scatterplot(x='Country', y='Trade Value', data=df, s=50, color='blue')
plt.title('Scatter Plot of Trade Value by Country')
plt.xlabel('Country')
plt.ylabel('Trade Value (in millions)')
plt.xticks(rotation=90)
plt.show()

# Explanation: This scatter plot visualizes the Trade Value for each country, highlighting the variations in Trade Value across different countries.
```
