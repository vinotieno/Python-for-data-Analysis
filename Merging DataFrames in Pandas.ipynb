# Merging DataFrames in Pandas



In this series we will be walking through everything you need to know to get started in Pandas! In this video, we learn about Merging, Joining, and Concatenating Dataframes in Pandas.


```python
import pandas as pd

df1 = pd.read_csv(r'C:\Users\ADMIN\Downloads\LOTR merge join concatenate .csv')

df1
```

```python
import pandas as pd

df2 = pd.read_csv(r'C:\Users\ADMIN\Downloads\LOTR 2 merge join concatenate.csv')

df2
```

### Merge

###### Inner Join 

```python
df1.merge(df2)


inner join----> where keys are the same 
```

```python
df1.merge(df2, how = 'inner')
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID' )
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

#### Outer join 


it gives us all the values regardless of if they are the same 

```python
df1.merge(df2, how = 'outer')
```

#### Left join

its going to take everything from the left table 1/or df(df1) and if there is any overlap it will also pull the overlapped or whatever we're able to merge/compare similarity on from table 2 or df2  


```python
df1.merge(df2, how = 'left')
```

#### Right join 

we are looking at the right hand ie table 2 and compare it to tabel 1 and merge them




```python
df1.merge(df2, how = 'right')
```

#### Cross


It takes each values from the left df( table 1) and compares it to each value in the right data freame(table 2)




```python
df1.merge(df2, how = 'cross')
```

### Join 

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left',rsuffix = '_Right')
```

```python
#### creating our index
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')

df4
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right')
df4
```

#### Concatenate

putting one df on top of the other 



```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner')
```

```python
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```

```python
#### Append 


It is used to append rows from one df to the end of another df  and then we return that new df 

```
